````python
x = 7
y = 10

print(x == y)
print(x != y)
print(x > y)
print(x < y)
print(x >= 7)
print(y <= 7)
````

````python
#Bitwise Operators


a = 5  # 0101
b = 3  # 0011

print(a & b)
print(a | b)
print(a ^ b)
print(~a)
print(a << 1)
print(a >> 1)

````
