```python


#\n → newline

#\t → tab

#\" → double quote inside a string

#\\ → backslash itself



# Newline
print("Hello\nWorld")

# Tab
print("Hello\tWorld")

# Including quotes inside string
print("He said, \"Python is fun!\"")

# Backslash
print("Path: C:\\Users\\Rahim")

# Carriage return
print("Hello\rWorld")

# Octal and Hex representation
print("\101")
print("\x41")
```
