## ** **accessing** items**

```python



fruits = ["apple", "banana", "cherry", "mango", "orange"]




print("First item:", fruits[0])            #  Access first item


print("Last item:", fruits[-1])              #  Access last item


print("Second item:", fruits[1])             # Access using index


print("Items from index 1 to 3:", fruits[1:4])    #Access a range (slice)


print("First 3 items:", fruits[:3])    #  Access from beginning to certain index
```

# **updating items**

```python


numbers = [10, 20, 30, 40, 50]




#  Update single item
numbers[0] = 100
print("After updating first item:", numbers)

#  Update last item
numbers[-1] = 500
print("After updating last item:", numbers)

#  Update multiple items at once
numbers[1:3] = [200, 300]
print("After updating index 1 and 2:", numbers)

#  Update using math operation
numbers[2] += 50
print("After adding 50 to index 2:", numbers)

#  Replace part of list with longer list
numbers[3:] = [400, 450, 600]
print("After replacing from index 3:", numbers)
```

# ` iterating`





```python

# -------------------------------

colors = ["red", "blue", "green", "yellow", "black"]


#  Simple for loop

for c in colors:
    print("Color:", c)

#  Using index with range()

for i in range(len(colors)):
    print(f"Index {i} has {colors[i]}")

#  Using enumerate() (index + value)

for i, c in enumerate(colors):
    print(f"At position {i}, color is {c}")

#  Iterating in reverse

for c in reversed(colors):
    print("Reverse order:", c)

# Inline iteration with condition

[print("Long name:", c) for c in colors if len(c) > 4]
```

## ** comprehensive**

```python



nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]



#  Square of numbers

squares = [n**2 for n in nums]
print("Squares:", squares)


#  Even numbers

evens = [n for n in nums if n % 2 == 0]
print("Even numbers:", evens)

# Odd numbers

odds = [n for n in nums if n % 2 != 0]
print("Odd numbers:", odds)

#  Multiply by 2
doubled = [n*2 for n in nums]
print("Doubled numbers:", doubled)


# Length of words

words = ["apple", "banana", "kiwi", "mango"]
lengths = [len(w) for w in words]
print("Word lengths:", lengths)
```

## manipulating *data*

```python



animals = ["cat", "dog", "rabbit"]



#  Add element at end

animals.append("elephant")
print("After append:", animals)


# Insert at specific position

animals.insert(1, "tiger")
print("After insert:", animals)


# Remove by value

animals.remove("dog")
print("After remove:", animals)

#  Pop element by index

popped = animals.pop(2)
print("After pop:", animals)
print("Popped element:", popped)

#  Sort and Reverse

animals.sort()
print("After sort:", animals)
animals.reverse()
print("After reverse:", animals)
```
