# **set operator**

```python
#union
set1={1,2,3,4,5}
set2={4,5,6,7}
print(set1 | set2)
```

```python
#union
set1={1,2,3,4,5}
set2={4,5,6,7}
print(set1. union(set2))
```

```python
#intersection
set1={1,2,3,4,5}
set2={4,5,6,7}
print(set1 & set2)
```

```python
#intersection
set1={1,2,3,4,5}
set2={4,5,6,7}
print(set1.intersection (set2))
```

```python
#difference

set1={1,2,3,4,5}
set2={4,5,6,7}
print(set1 - set2)
```

```python
#difference

set1={1,2,3,4,5}
set2={4,5,6,7}
print(set1. difference (set2))
```
