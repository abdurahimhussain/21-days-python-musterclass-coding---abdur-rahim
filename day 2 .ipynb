# scope of **variable**

```python
#local variable
def func ():
  x=10
  print(x)


  func()
print(x)
```

```python
#globally
   x= 20

def func():
      x=10
  print(x)


  func()
  print(x)
```

```python

#Enclosed Scope (Nested Function)


def outer():
    a = "outer variable"

    def inner():
        print("Accessing from inner:", a)  # enclosed scope

    inner()

outer()
```

```python
#Built-in Scope

print(len("Hello"))   # len is from built-in scope
```

```python
#Using global Keyword


count = 0

def update_count():
    global count
    count += 1
    print("Inside function:", count)

update_count()
print("Outside function:", count)
```

```python
#Using nonlocal Keyword


def outer():
    num = 10
    def inner():
        nonlocal num
        num += 5
        print("Inside inner:", num)
    inner()
    print("Inside outer:", num)

outer()
```
